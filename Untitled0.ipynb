```python

!pip install PyPDF2

import os
import PyPDF2
import pandas as pd
import re
from google.colab import files

#The dictionary of categories
categories_map = {
    'Groceries': ['trader joe', 'butcherie', 'costco', 'supermarket', 'grocery'],
    'Dining Out': ['pizza', 'stoked', 'veggie crust', 'bakery', 'mcdonalds', 'tst*', 'cafe', 'tavern', 'bakey'],
    'Transport': ['u-haul', 'sunoco', 'uber', 'mbta', 'gulf market', 'cumberland'],
    'Leisure': ['amc', 'netflix', 'google one', 'apple.com', 'levi'],
    'Health': ['pharmacy', 'cvs', 'doctor'],
    'Bills/Services': ['eversource', 'barbers', 'zelle', 'venmo', 'mobile banking payment']
}

def smart_categorize(description):
    description = description.lower()
    for category, keywords in categories_map.items():
        for word in keywords:
            if word in description:
                return category
    return 'Other'

def clean_amount(amt_str):
    try:
        return float(amt_str.replace('$', '').replace(',', '').replace('-', ''))
    except:
        return 0.0

print("The library has been installed and everything is ready!")
```

```python
all_expenses = []
folder_path = '/content/'

for filename in os.listdir(folder_path):
    if filename.endswith('.pdf'):
        print(f"Processing: {filename}")
        with open(os.path.join(folder_path, filename), 'rb') as pdf_file:
            reader = PyPDF2.PdfReader(pdf_file)
            for page in reader.pages:
                text = page.extract_text()
                for line in text.split('\n'):
                    #Searching for expense lines (date and minus)
                    if re.search(r'\d{2}/\d{2}/\d{2}', line) and '-' in line:
                        parts = line.split()
                        amount = clean_amount(parts[-1])
                        description = " ".join(parts[1:-1])
                        category = smart_categorize(description)

                        all_expenses.append({
                            'Source_File': filename,
                            'Category': category,
                            'Amount': amount,
                            'Description': description[:30]
                        })

df_all = pd.DataFrame(all_expenses)
print(f"Done! Collected {len(df_all)} transactions.")
```

```python
# 1. Create a summary by category
final_summary = df_all.groupby('Category')['Amount'].sum().reset_index()

# 2. Displaying a pie chart
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 7))
plt.pie(final_summary['Amount'], labels=final_summary['Category'], autopct='%1.1f%%', startangle=140)
plt.title('Family Spending Summary - Boston 2026')
plt.show()

# 3. Download to Excel
final_summary.to_csv('family_summary.csv', index=False)
files.download('family_summary.csv')

print("Your organized Excel is ready!")
```
